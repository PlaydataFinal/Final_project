```python
import pandas as pd

# CSV 파일 경로 설정
# file_path = r'C:\Users\Playdata\Desktop\Project\data\file_name.csv'

# CSV 파일 불러오기
df1 = pd.read_csv('data/kakaoapi_tour_kakao.csv')
df2 = pd.read_csv('data/tour_index_vectorized_list.csv')
```

```python
df1
```

```python
# 불필요칼럼 제거
df1 = df1[['id','Name']]
df2 = df2[['Name', 'Tag','Hits','embedding']]

```

```python
import pandas as pd

# df1과 df2에서 각각 'Name'을 기준으로 중복 제거
df1_unique = df1.drop_duplicates(subset='Name')
df2_unique = df2.drop_duplicates(subset='Name')

# df1과 df2의 중복 제거된 Name을 기준으로 외부 조인
merged_df = pd.merge(df1_unique, df2_unique, on='Name', how='outer')

# 중복된 값 제거
merged_df.drop_duplicates(subset='Name', inplace=True)

# 결과 출력
print(merged_df)
```

```python
merged_df.dropna(inplace=True)
merged_df
```

```python
merged_df['id'] = merged_df['id'].astype('int64')
```

```python
# id 오름차순으로 DataFrame 정렬
merged_df = merged_df.sort_values(by='id')
```

```python
web_recommend_df = merged_df.copy()
```

```python
web_recommend_df.reset_index(drop=True, inplace=True)
```

```python
web_recommend_df.dtypes
```

```python
from sentence_transformers import SentenceTransformer
import re

# 태그 칼럼 임베딩
encoder = SentenceTransformer('jhgan/ko-sroberta-multitask')

# 정규표현식을 사용하여 해시태그 제거 및 단어만 추출하는 함수
def extract_words(text):
    # 해시태그를 제거하고 단어만 추출
    words = re.findall(r'\w+', text)
    return ' '.join(words)  # 단어들을 공백으로 구분하여 문자열로 반환

# Tag 칼럼의 데이터에 대해 정규표현식을 적용하여 해시태그를 제거하고 단어만을 추출하여 임베딩 처리
web_recommend_df['tag_embedding'] = web_recommend_df['Tag'].map(lambda x: list(encoder.encode(extract_words(x))))

```

```python
import numpy as np
# 문자열을 배열로 변환하는 함수 정의
def parse_embedding(embedding_str):
    return np.array([float(val) for val in embedding_str.strip('[]').split(', ')])

# 'embedding' 열의 값을 배열로 변환
web_recommend_df['embedding'] = web_recommend_df['embedding'].apply(parse_embedding)

# 'tag_embedding' 열의 값을 배열로 변환
web_recommend_df['tag_embedding'] = web_recommend_df['tag_embedding'].apply(parse_embedding)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 임베딩 값 추출
embeddings = np.array([np.array(embedding, dtype=float) for embedding in web_recommend_df['embedding']])
tag_embeddings = np.array([np.array(embedding, dtype=float) for embedding in web_recommend_df['tag_embedding']])

# 임베딩 유사도 계산
similarities = cosine_similarity(embeddings)

# 상위 40개의 유사한 관광지 인덱스 추출
top_40_indices = np.argsort(similarities, axis=1)[:, -40:-1][:, ::-1]

# 각 관광지별로 태그 임베딩 간의 유사도 계산하여 상위 3개 선택
for i in range(len(web_recommend_df)):
    tag_similarities = cosine_similarity([tag_embeddings[i]], tag_embeddings[top_40_indices[i]])
    top_3_tag_indices = np.argsort(tag_similarities)[:, -4:-1][:, ::-1][0]
    
    # 상위 3개의 인덱스를 저장하고 데이터 형식을 int64로 변환
    web_recommend_df.at[i, 'rec1'] = int(web_recommend_df.at[top_40_indices[i, top_3_tag_indices[0]], 'id'])
    web_recommend_df.at[i, 'rec2'] = int(web_recommend_df.at[top_40_indices[i, top_3_tag_indices[1]], 'id'])
    web_recommend_df.at[i, 'rec3'] = int(web_recommend_df.at[top_40_indices[i, top_3_tag_indices[2]], 'id'])
    
    # 상위 3개의 유사도 값 추출
    top_3_tag_similarities = np.sort(tag_similarities[0])[-4:-1][::-1]
    
    # 유사도 값을 'sim1', 'sim2', 'sim3' 열에 저장
    web_recommend_df.at[i, 'sim1'] = top_3_tag_similarities[0]
    web_recommend_df.at[i, 'sim2'] = top_3_tag_similarities[1]
    web_recommend_df.at[i, 'sim3'] = top_3_tag_similarities[2]

# 'rec1', 'rec2', 'rec3' 열의 데이터 형식을 int64로 변환
web_recommend_df['rec1'] = web_recommend_df['rec1'].astype('int64')
web_recommend_df['rec2'] = web_recommend_df['rec2'].astype('int64')
web_recommend_df['rec3'] = web_recommend_df['rec3'].astype('int64')
```
